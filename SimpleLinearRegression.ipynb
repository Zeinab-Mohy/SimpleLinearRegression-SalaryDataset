```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

# Load DataSet

```python
#load dataSet
data=pd.read_csv('/kaggle/input/salary-dataset-simple-linear-regression/Salary_dataset.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
x=data['YearsExperience']
y=data['Salary']
```

```python
plt.scatter(x,y)
plt.xlabel("Salary")
plt.ylabel("YearOfExperiences")
plt.show()
```

# Train DataSet

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,train_size=0.8,random_state=101)
```

```python
X_train = np.array(X_train).reshape(-1, 1)
X_test = np.array(X_test).reshape(-1, 1)
```

# Apply model

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
y_predict
```

```python
plt.scatter(X_train,y_train)
plt.plot(X_train,model.predict(X_train))
plt.show()
```

```python
CoefficientOfDetermination=r2_score(y_test,y_predict)
CoefficientOfDetermination
print("Accurracy : ",CoefficientOfDetermination)
```
